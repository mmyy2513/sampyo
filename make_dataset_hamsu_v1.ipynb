# import

```python
import cv2
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

# crop position test

```python
dataset_name = "./origin_data_water/FM_2.4_2.5%/"

FM_list = os.listdir(dataset_name)

# for one_folder in folder_list:
    
w = 1500
h = 1700
start_point = (350, 350)
end_point = (start_point[0] + w, start_point[1] + h)

img = cv2.imread(dataset_name + FM_list[0], cv2.IMREAD_COLOR)
img = cv2.rectangle(img, start_point, end_point, (0,255,0), 10)

plt.figure(figsize=(8,8))
plt.imshow(img, cmap="brg")
```

# FM 2.4(2.5%) crop

```python
dataset_name = "./origin_data_water/FM_2.4_2.5%/"
crop_dataset_name = "./crop_dataset_water/FM_2.4_2.5%/"

w = 1500
h = 1700
start_point = (350, 350)
end_point = (start_point[0] + w, start_point[1] + h)


img_list = os.listdir(dataset_name)

for one_img in img_list:
    if one_img == "Thumbs.db":
        continue
    join_img_path = os.path.join(dataset_name, one_img)
    img = cv2.imread(join_img_path, cv2.IMREAD_COLOR)
    crop_img = img[start_point[1]:end_point[1],start_point[0]:end_point[0],:]
    cv2.imwrite(crop_dataset_name + one_img, crop_img)
```
